#Question 1 : Explain the basic structure of an HTML page. Why is it important to follow this Structure?

- HTML stands for hyper text markup language. It is the standard markup language used to create the structure and content of web pages displayed in a web browser.

```html

<!DOCTYPE html>
<html>
<head>
    <meta charset="UTF-8">
    <title>Page Title Goes Here</title>
    </head>
<body>
    <h1>Main Heading</h1>
    <p>A paragraph of content.</p>
</body>
</html>

 - Explanation:
 . <DOCTYPE html> - This is the document type declaration. It tells the browser that the document is an HTML5 page.
  . <html>: This is the root element of the page. All other elements (except the <!DOCTYPE html>) must be nested inside this tag. It often includes an attribute specifying the language (e.g., <html lang="en">).
 .  <head>: This section contains metadata about the HTML document that is not displayed directly on the web page. This includes:
 . <meta charset="UTF-8">: Specifies the character encoding for the document (UTF-8 is standard).
 . <title>: Sets the title that appears in the browser tab or window title bar.
 . <body>: This section contains the visible content of the page. Everything a user sees and interacts with (headings, paragraphs, images, links, etc.) is placed within this tag.

   . Importance of following the Structure-

    .Browser Rendering-Ensures correct and consistent display

    .Accessibility-Makes content readable by screen readers and assistive tech.

    .SEO-Helps search engines understand and rank the page.


Question 2: You’re building a small personal profile page. Which HTML tags will youuse to highlight your name as a heading, describe yourself in a paragraph, and emphasize a key skill?

```html
<!DOCTYPE html>
<html>
<head>
    <meta charset="UTF-8">
    <title>My Web Page</title>
    </head>
<body>
    <h1>Amitesh Pandey</h1>
    <p>I am Amitesh Pandey, I am from Prayagraj, UP, and I am currently in Delhi doing college. My college name is Amity University, where I am studying BCA, and I currently do video editing work.</p>
    <p>My Favourite work is <em>Video Editing</em></p>
</body>
</html>



             

# Question 3: Look at the code below. Can you find and fix the mistake?
```html
 <!-- This is an HTML comment
<h1>Welcome to my website</h1>
-The comment tag is not closed properly, it should be ended with -->

<!-- This is an HTML comment-->
<h1>Welcome to my website</h1>




#Question 4: You’re creating a form that asks the user to enter their name, select their country from a list, and click a button to submit. Write the HTML code for this.

 -This the html code for the form which asks the user to enter their name and name select their country -

 ```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
</head>
<body>
    <form >
        <label for="name">Name</label>
        <input type="text" name=username"" id="name" placeholder="Enter Your Name" required>
    </br>
</br>
    <label for="Country">Country</label>
    <select name="country" id="country">
        <option value="india">India</option>
        <option value="pakistan">Pakistan</option>
        <option value="unitedkingdom">United Kingdom</option>
        <option value="unitedstates">United States</option>
    </select>
        </br>
</br>
    <button type="submit">Submit Form</button>

    </form>
</body>
</html>


# Question 5: Write short HTML code to show your favorite fruits in a bulleted list and your top 3 programming languages in a numbered list.
```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>My Web Page</title>
</head>
<body>
    <h2>My Favourite Fruits</h2>
    <ul>
        <li>Mango</li>
        <li>Apple</li>
        <li>Banana</li>
        <li>Orange</li>
        <li>Papaya</li>
    </ul>
    <h2>Top 3 Programming Language </h2>
        <ol>
            <li>Python</li>
            <li>Java</li>
            <li>C/C++</li>
        </ol>
</body>
</html>

# Question 6: Ravi wants to leave a note in his HTML file so that only other developers can see it, not the users on the website. How should he do this? Why are such notes useful?

 - Ravi should use an HTML comment tag.
 - Why Comments are Useful?

    . Documentation: Explains complex or non-obvious code sections.
    
    . Organization: Marks sections


# Question 7: Here’s a form snippet. Can you spot the error and rewrite the correct
version?
```html
<form>
 <input name="email">
 <select>
 <option>India</option>
 </select>
 <button>Send</buton>
</form>

 - Key Corrections in this form snippet are-
 . </select>: Added the mandatory closing tag for the <select> element.

 .value="Send": Added the value attribute to the submit input to display the desired text "Send" on the button.

 . Context: Wrapped the elements in a basic <form> tag and added a <label> for better accessibility and user experience.

  -Correct Version of this Snippet is -
  <form action="/submit-form">
    <label for="country">Country:</label>
    <select id="country" name="country">
        <option value="india">India</option>
        <option value="usa">USA</option>
        <option value="canada">Canada</option>
    </select>
    <br>
    <input type="submit" value="Send">
</form>
